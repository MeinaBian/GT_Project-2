```python
counter = 0
recipe_id = []
recipe_title = []
recipe_sodium = []
recipe_sugar = []

# recipe_calories = []
# recipe_protein = []
# recipe_fat = []
# recipe_carbs = []

while counter < 100:
    url = "https://api.spoonacular.com/recipes/findByNutrients?maxSugar=25&maxSodium=100&number=500&random=true&apiKey=a21364208c9d43719315d65c34faa384"
    query_url = f'{url}&apiKey={api_Key}'
    response = requests.get(query_url).json()
    for i in range(len(response)):
        try:
            recipe_id.append(response[i]['id'])
        except KeyError:
            recipe_id.append('null')
        recipe_title.append(response[i]['title'])
        recipe_sodium.append(response[i]['sodium'])
        recipe_sugar.append(response[i]['sugar'])
#       recipe_calories.append(response[0]['calories'])
#       recipe_protein.append(response[0]['protein'])
#       recipe_fat.append(response[0]['fat'])
#       recipe_carbs.append(response[0]['carbs'])
    counter += 1
```
